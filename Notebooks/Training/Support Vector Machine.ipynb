```python
import os
import numpy as np
import pandas as pd
import csv
from sklearn import svm
```

```python
file = pd.read_csv("test features.csv")
#low level features
```

```python
X = file.iloc[:, 1:28].values
y = file.iloc[:, 29].values
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
classifier = svm.SVC() #one-against-one
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
score = classifier.score(X_test, y_test)
coefficient = classifier.dual_coef_
print(score)
print(coefficient)
```

```python
nuclassifier = svm.NuSVC(nu=0.6) #one-against-one
nuclassifier.fit(X_train, y_train)
y_pred = nuclassifier.predict(X_test)
score = nuclassifier.score(X_test, y_test)
coefficient = nuclassifier.dual_coef_
print(score)
print(coefficient)
```

```python
linear_classifier = svm.LinearSVC(max_iter=2500) #one-against-many
linear_classifier.fit(X_train, y_train)
y_pred = linear_classifier.predict(X_test)
score = linear_classifier.score(X_test, y_test)
coefficient = linear_classifier.coef_
print(score)
print(coefficient)
```
