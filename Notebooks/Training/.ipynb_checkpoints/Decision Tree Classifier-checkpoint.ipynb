```python
import os
import numpy as np
import pandas as pd
import csv
from sklearn import tree
```

```python
file = pd.read_csv("test features.csv")
```

```python
X = file.iloc[:, 1:28].values
y = file.iloc[:, 29].values
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
#remove cell if using high level features
```

```python
from sklearn import tree

classifier = tree.DecisionTreeClassifier()
classifier.fit(X_train, y_train)
y_pred = classifier.predict(X_test)
score = classifier.score(X_test, y_test)
print(score)
print(y_pred)
```

```python
import matplotlib.pyplot as plt

classes = ['Brave', 'Cheerful', 'Fearful', 'Love']
features = list(file)
fig, ax = plt.subplots(figsize=(20, 20))
tree = tree.plot_tree(classifier, class_names=classes, feature_names=features[1:28], fontsize=12, ax=ax)
plt.show()
```
